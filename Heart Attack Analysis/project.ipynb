# Data Preprocessing

```python
df.rename(columns={"98.6":"o2Saturation","trtbps":"blood_pressure",
                   "chol":"Cholestoral","fbs":"fasting_blood_sugar","thalachh":"max_heart_rate",
                   "exng":"exercise"},inplace=True)
```

```python
print('Number of rows are :',df.shape[0], ',and number of columns are :',df.shape[1])
```

```python
df.output.value_counts()
print('\n 0 less chance of heart attack =',df.output.value_counts()[0],"\n",
      '1 more chance of heart attack=',df.output.value_counts()[1])
```

Sorting the high and low sugar level count.

```python
df.fasting_blood_sugar.value_counts()
print(
    '\n Value 0: fasting blood sugar > 120 mg/dl) (0 = false)',df.fasting_blood_sugar.value_counts()[0],"\n",
    'Value 1: fasting blood sugar > 120 mg/dl) (1 = true)',df.fasting_blood_sugar.value_counts()[1],"\n",)
```

Sorting the sex count on the Man and Woman.

```python
df.sex.value_counts()
print('\n 1 Man =',df.sex.value_counts()[0],"\n",
      '0 Woman=',df.sex.value_counts()[1])
```

```python
df.loc[(df['sex'] ==1), 'sex'] = 'Man'
df.loc[(df['sex'] ==0), 'sex'] = 'Woman'

df.loc[(df['output'] ==1), 'output'] = 'High'
df.loc[(df['output'] ==0), 'output'] = 'Low'

df.head()
```

```python
df.cp.value_counts()
print(
    '\n Value 0: typical angina  =',df.cp.value_counts()[0],"\n",
      'Value 1: atypical angina =',df.cp.value_counts()[1],"\n",
      'Value 2: non-anginal pain=',df.cp.value_counts()[2],"\n",
      'Value 3: asymptomatic    =',df.cp.value_counts()[3])
```
